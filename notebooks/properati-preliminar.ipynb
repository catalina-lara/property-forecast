```python
import folium
from folium.plugins import MarkerCluster

# Crear el mapa centrado en Montevideo
m = folium.Map(location=[-34.9, -56.2], zoom_start=12)

# Agrupar los puntos
marker_cluster = MarkerCluster().add_to(m)

# Agregar los puntos válidos
for i, row in data_geo.iterrows():
    folium.Marker(
        location=[row['lat'], row['lon']]
    ).add_to(marker_cluster)

m
```

```python
data_uruguay_mv_vt['start_date'] = pd.to_datetime(data_uruguay_mv_vt['start_date'], errors='coerce')
data_uruguay_mv_vt['end_date'] = pd.to_datetime(data_uruguay_mv_vt['end_date'], errors='coerce')
```

```python
data_uruguay_mv_vt[['start_date', 'end_date']].agg(['min', 'max'])
```

```python
# Agrupar por mes de la fecha de inicio
registros_por_mes = (
    data_uruguay_mv_vt
    .groupby(data_uruguay_mv_vt['start_date'].dt.to_period('M'))
    .size()
)

# Convertir a índice tipo datetime para graficar fácilmente
registros_por_mes.index = registros_por_mes.index.to_timestamp()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))
plt.plot(registros_por_mes.index, registros_por_mes.values, marker='o')
plt.title('Cantidad de registros por mes (start_date)')
plt.xlabel('Mes')
plt.ylabel('Cantidad de registros')
plt.grid(True)
plt.show()
```
